# Pie Charts 

```python
colors = ['#cc0051', '#e6005b', '#ff0065', '#f6546a']
```

```python
labels = 'Pizza', 'Tacos', 'Pasta', 'Chinese'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Dogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
labels = 'Frogs', 'Dogs', 'Cats', 'Birds'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Dogs')
colors = ('#cc0051', '#e6005b', '#ff0065', '#f6546a')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors=colors, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
display(HTML('<img src="https://random-ize.com/coin-flip/us-quarter/us-quarter-front.jpg">'))
```

```python
heads = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-front.jpg'
tails = 'https://random-ize.com/coin-flip/us-quarter/us-quarter-back.jpg'
```

```python
def show_img(img):
  display(HTML('<img src=' + img + '>'))
```

```python
show_img(heads)
```

```python
show_img(tails)
```

```python
def flip_coin():
  if coin():
      show_img(heads)
  else: 
      show_img(tails)   
```

```python
flip_coin()
```

```python
flip_coin()
```
